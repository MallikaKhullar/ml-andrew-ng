## Labels, Title & limits

```python
pl.xlabel('Injuries')
pl.ylabel('Sport')
pl.title("SPORTS")
pl.xlim(-2000, 100)
pl.ylim(-2000, 100)
pl.plot([21,56,3], [1, 80, 45], 'r*')
pl.show()
```

## Multiple Plots

```python
pl.xlabel('Injuries')
pl.ylabel('Sport')
pl.title("SPORTS")
pl.plot([21,56,3], [1, 80, 45], 'r*')
pl.plot([20,506,3], [100, 80, 405], 'b+')
pl.show()
```

```python
pl.xlabel('Injuries')
pl.ylabel('Sport')
pl.title("SPORTS")
pl.plot([21,56,3], [21, 80, 45], 'r--')
pl.plot([20,506,3], [0, 80, 15], 'b-.')
pl.plot([100,20,50], 'y-') #plot using x = 0
pl.show()
```

## Legends

```python
plot1, = pl.plot([1,10], [2,50], 'r') #commas are imp
plot2, = pl.plot([12,1], [20,100], 'go') #commas are important
pl.legend([plot1, plot2], ("G", "E")) 
pl.show()
```
